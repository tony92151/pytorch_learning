```python
import torch
from torch import nn, optim
import torch.nn.functional as F
import matplotlib.pyplot as plt
import sys
import time
import numpy as np
import math
import pandas as pd
from PIL import Image, ImageOps, ImageFilter
from datetime import datetime
from torch.autograd import Variable
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
from torchvision import datasets, models, transforms
import random
import datetime
import os

from sklearn import preprocessing 
from sklearn.model_selection import KFold
print(torch.__version__)
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print(device)
```

```python
def onehot(length, num):
    d = np.zeros(length)
    d[num] = 1
    return d
```

```python
def dataread(data_dir):
    datt = []
    filenames = os.listdir(data_dir)
    file = [os.path.join(data_dir, f) for f in filenames if f.endswith('.jpg')]
    labels2 = [int(os.path.split(filename)[-1][0]) for filename in file]

    #print(len(file),len(labels2))
    for i in range(len(file)):
        datt.append([file[i],onehot(6,labels2[i])])
        #print(len([file[i],onehot(6,labels2[i])]))
    return datt
```

```python
class SIGNSDataset(Dataset):
    def __init__(self, data_dir, transform):
        self.filenames = os.listdir(data_dir)
        self.filenames = [os.path.join(data_dir, f) for f in self.filenames if f.endswith('.jpg')]

        #self.labels = [int(os.path.split(filename)[-1][0]) for filename in self.filenames]
        self.labels2 = [int(os.path.split(filename)[-1][0]) for filename in self.filenames]
        #print(self.labels2)
        self.labels = np.array([onehot(6,i) for i in self.labels2])
        #self.filenames = data_dir[:,0]
        #self.labels = data_dir[:,1]
        self.transform = transform

    def __len__(self):
        # return size of dataset
        return len(self.filenames)

    def __getitem__(self, idx):
        image = Image.open(self.filenames[idx])  # PIL image
        #image = Image.open(self.filenames[idx]).convert('LA')  # PIL image
        #image = ImageOps.equalize(image, mask = None)
        image = self.transform(image)
        return image,self.labels[idx]
```

```python
imgsize = 64
```

```python
train_data_path = "SIGNS_dataset/train_signs/"
test_data_path = "SIGNS_dataset/test_signs/"
```

```python
train_transformer = transforms.Compose([
    transforms.Resize((imgsize,imgsize)),
    transforms.RandomHorizontalFlip(),  # randomly flip image horizontally
    transforms.ToTensor(),  # transform it into a torch tensor
#     transforms.Normalize(mean=[0.485, 0.456, 0.406],
#                          std=[0.229, 0.224, 0.225])
    transforms.Normalize(mean = (0.5, 0.5, 0.5), std = (0.5, 0.5, 0.5))
    ])

test_transformer = transforms.Compose([
    transforms.Resize((imgsize,imgsize)),
    transforms.RandomHorizontalFlip(),  # randomly flip image horizontally
    transforms.ToTensor(),  # transform it into a torch tensor
    transforms.Normalize(mean = (0.5, 0.5, 0.5), std = (0.5, 0.5, 0.5))
    ])
```

```python
traindata = SIGNSDataset(train_data_path, train_transformer)
testdata = SIGNSDataset(test_data_path, test_transformer)

dl = DataLoader(traindata, batch_size=32, shuffle=True,
                                        num_workers=4,
                                        pin_memory=True)

tl = DataLoader(testdata, batch_size=32, shuffle=False,
                                        num_workers=4,
                                        pin_memory=True)
```

```python
print("Train data: ",len(traindata))
print("Test data: ",len(testdata))
```

```python
index = 8
plt.imshow(transforms.ToPILImage()(tl.dataset[index][0]).convert('RGB'))

print(np.argmax(tl.dataset.labels[index]))
print(tl.dataset.labels[index])
```

```python
model = models.mobilenet_v2()
model.classifier = nn.Sequential(
          #nn.Dropout(p=0.1, inplace=False),
          nn.Linear(in_features=1280, out_features=64, bias=True),
          #nn.Dropout(p=0.2, inplace=False),
          nn.Linear(in_features=64, out_features=6))
model.to(device)
```

```python
#optimizer = torch.optim.Adam(model.parameters(),lr=0.01)
optimizer = torch.optim.SGD(model.parameters(),lr=0.001)
loss_func = nn.CrossEntropyLoss() 
```

```python
train_loss = []
test_loss = []
test_acc = []
```

```python
def train(epoch):
    model.train()
    for i, (train_batch, labels_batch) in enumerate(dl):
        labels_batch = np.argmax(labels_batch,axis=1)
        train_batch, labels_batch = Variable(train_batch).cuda(), Variable(labels_batch).cuda()
        
        output_batch = model(train_batch)
        loss = loss_func(output_batch, labels_batch.long())
        optimizer.zero_grad()
        loss.backward()

        optimizer.step()
        torch.cuda.empty_cache()
        if i % 8 == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(epoch, 
                    i * len(train_batch), len(dl.dataset),100. * i / len(dl), loss.data.item()))
    train_loss.append(loss)
    print("")
```

```python
#set_printoptions(formatter=dict(float=lambda t:"%14.7f" % t))
def test():
    acc = 0
    los = 0
    model.eval()
    for i, (test_batch, labels_batch) in enumerate(tl):
        labels_batch = np.argmax(labels_batch,axis=1)
        test_batch, labels_batch = Variable(test_batch).cuda(), Variable(labels_batch).cuda()
        with torch.no_grad():
            output_batch = model(test_batch)
        
        loss = loss_func(output_batch, labels_batch.long())
        los += loss
        #test_loss.append(loss)
        
        output_batch = np.argmax(output_batch.cpu().numpy(),axis=1)
        labels_batch = labels_batch.cpu().numpy()
        #print(output_batch.shape)
        #print(labels_batch.shape)
        
        #print((output_batch==labels_batch))
        acc += (output_batch==labels_batch).sum()
        #print(acc)
        
        torch.cuda.empty_cache()
    Acc = acc/len(tl.dataset.labels)
    Loss = los/len(tl.dataset.labels)
    test_loss.append(Loss)
    test_acc.append(Acc)
    print("Test loss: {:.6f} \t Acc: [{}/{}] {:.4}".format(Loss,
                                                           acc,
                                                           len(tl.dataset.labels),
                                                           Acc))
    print("")
```

```python
Epoch = 50
for epoch in range(Epoch):
    if epoch < (20):
        optimizer = torch.optim.SGD(model.parameters(),lr=0.01)
    elif epoch >= (20) and epoch < (30) :
        optimizer = torch.optim.SGD(model.parameters(),lr=0.001)
    else:
        optimizer = torch.optim.SGD(model.parameters(),lr=0.0008)
    train(epoch)
    test()
```
