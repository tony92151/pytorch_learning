```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import torch.utils.data as Data
import torchvision
import matplotlib.pyplot as plt
import numpy as np
```

```python
print('version :' , torch.__version__)
print('cuda :' , torch.cuda.is_available())
print('cudnn :' , torch.backends.cudnn.enabled)
```

```python
EPOCH = 5
BATCH_SIZE = 50
LR = 0.001
DOWNLOAD_MNIST = True
```

```python
#torch.cuda.current_device()
#torch.cuda.current_stream()
torch.cuda.empty_cache()
```

```python
train_data = torchvision.datasets.MNIST(
    root = './mnist',
    train = True,
    transform=torchvision.transforms.ToTensor(), #改成torch可讀
    download=DOWNLOAD_MNIST,
)
```

```python
print(len(train_data.train_data))
print(len(train_data.train_labels))
```

```python
for i in range(10):    
    img = np.asarray(train_data.train_data[i])
    plt.imshow(img, cmap='gray')
    plt.title('%i' % train_data.train_labels[i])
    plt.show()
```

```python
train_loader = Data.DataLoader(dataset=train_data, batch_size=BATCH_SIZE, shuffle=True)
#print(train_loader.dataset)
test_data = torchvision.datasets.MNIST(root='./mnist/', train=False)
print(test_data)
print(type(test_data.test_data))

#print(len(test_data.test_data))
test_x = Variable(torch.unsqueeze(test_data.test_data, dim=1)).type(torch.FloatTensor)[:3000].cuda()/255.   # shape from (2000, 28, 28) to (2000, 1, 28, 28), value in range(0,1)
test_y = test_data.test_labels[:3000].cuda()
```

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Sequential(         # input shape (1, 28, 28)
            nn.Conv2d(                      # output shape (16, 24, 24)
                in_channels=1,              # input height, only 1 (0 or 1)
                out_channels=16,            # n_filters
                kernel_size=5,              # filter size
                stride=1,                   # filter movement/step
                padding=1,                  # if want same width and length of this image after con2d, padding=(kernel_size-1)/2 if stride=1
            ),                              
            nn.ReLU(),                      # activation
            nn.MaxPool2d(kernel_size=2),    # output shape (16, 12, 12)
        )
        self.conv2 = nn.Sequential(         # input shape (16, 12, 12)
            #nn.Conv2d(16, 32, 5, 1,1),     # output shape (32, 10, 10)
            nn.Conv2d(16, 32, 5, 1),     # output shape (32, 8, 8)
            nn.ReLU(),                      # activation
            nn.MaxPool2d(2),                # output shape (32, 5, 5)
            nn.MaxPool2d(2),                # output shape (32, 4, 4)
        )
        self.out = nn.Linear(32*4*4, 10)   # fully connected layer, output 10 classes

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = x.view(x.size(0), -1)           # flatten the output of conv2 to (batch_size, 32 * 7 * 7)
        output = self.out(x)
        return output    # return x for visualization
```

```python
cnn = CNN()
cnn.cuda()
print(cnn)
```

```python
optimizer = torch.optim.Adam(cnn.parameters(), lr=LR)   # optimize all cnn parameters
loss_func = nn.CrossEntropyLoss()                       # the target label is not one-hotted
```

```python
for epoch in range(EPOCH):
    for step, (x, y) in enumerate(train_loader):   # gives batch data, normalize x when iterate train_loader
        b_x = Variable(x).cuda()   # batch x
        b_y = Variable(y).cuda()   # batch y

        output= cnn(b_x)           # cnn output
        loss = loss_func(output, b_y)   # cross entropy loss
        optimizer.zero_grad()           # clear gradients for this training step
        loss.backward()                 # backpropagation, compute gradients
        optimizer.step()                # apply gradients

        if step % 100 == 0:
            #print(output)
            test_output = cnn(test_x)
            pred_y = torch.max(test_output, 1)[1].data.squeeze()
            accuracy = (pred_y == test_y).sum().item() / float(test_y.size(0))
            print('Epoch: ', epoch, '| train loss: %.4f' % loss.item(), '| test accuracy: %.2f' % accuracy)
print("Finish")
```

```python
# print 10 predictions from test data
test_output, _ = cnn(test_x[:10])
pred_y = torch.max(test_output, 1)[1].data.cpu().numpy().squeeze()
print(pred_y, 'prediction number')
print(test_y[:10].cpu().numpy(), 'real number')
```
